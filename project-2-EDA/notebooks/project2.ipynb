```python
df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month')['TotalPrice'].sum()
monthly_sales
```

```python
monthly_sales.plot(figsize=(10,5))

plt.title("Monthly Sales Trend")

plt.show()
```

```python
# Key Observations and Conclusions

## Order Status Analysis
- Cancelled orders (250) are slightly higher than delivered orders (231).
- A considerable number of orders do not reach successful completion.

## Payment Method Analysis
- Online payment is the most preferred payment method with 258 transactions.
- Cash, Credit Card, Debit Card, and Gift Card transactions are distributed fairly evenly.

## Referral Source Analysis
- Instagram generates the highest customer traffic with 259 referrals.
- Email and Google are also major traffic sources.

## Product Analysis
- Printer is the best-selling product with 181 sales.
- Tablet and Chair closely follow Printer in total sales count.

## Monthly Sales Trend
- Sales peaked during June 2024 with a total sales value of 68,068.54.
- Sales fluctuate across months, indicating seasonal purchasing behavior.

## Correlation Analysis
- UnitPrice and TotalPrice have a strong positive correlation (0.72).
- Quantity and TotalPrice also show a strong positive correlation (0.62).
- Quantity and ItemsInCart show a moderate positive correlation (0.65).

## Conclusion
The exploratory data analysis provided valuable insights into customer purchasing behavior, product performance, payment preferences, and sales trends. Online payments and Instagram referrals emerged as the most significant channels. Printer was identified as the best-selling product, while June 2024 recorded the highest sales. Correlation analysis revealed that TotalPrice is strongly influenced by both UnitPrice and Quantity. These findings can support better business decisions in marketing, inventory management, and customer engagement.
```
